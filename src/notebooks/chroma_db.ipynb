```python
import chromadb
# Initialize ChromaDB client
client = chromadb.Client()

# Create a new collection
if "my_collection" in client.list_collections():
    client.delete_collection(name="my_collection")
collection = client.create_collection(name="my_collection")

# Add some data to the collection
collection.add(
    documents=["This is document1", "This is document2","Another document"], # we handle tokenization, embedding, and indexing automatically. You can skip that and add your own embeddings as well
    ids=["doc1", "doc2","doc3"], # unique for each doc
)

print("ChromaDB setup complete.")

collection = client.get_collection(name="my_collection")

print(f"Docs in collection {collection.count()}")
```
